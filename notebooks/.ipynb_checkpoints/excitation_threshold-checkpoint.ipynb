# Computing the Excitation PCA Threshold (EPT)

We compute the excitation activations for each encoder depth and computed the ammount of principle components (PC) to explain 95% of the variance.

We find that number PC vs explained variance follows a double exponential decay trend:
    - The first few exponential components increase perforance at an exponential rate
    - Then when entering the regime of dimishing returns, modelled by a second exponential that decays faster than the first 


$|\mathbf{X} - \hat{Y}(\mathbf{P}_n)| \propto y(n) $
, so that
$y(n) = g \cdot e^{g \cdot n} - d \cdot e^{-d \cdot n} $, 

where:
- $\mathbf{X}$ is the value of the excitations accross a dataset
- $\hat{Y}(\mathbf{P}_n)$ is the prediction of $\mathbf{X}$ using $n$ of its principle components
- $y(n)$ is the model of explanied variance using $n$ principle components, $g, d \in \mathbb{R}$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import robust_scale
from sklearn.decomposition import PCA
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score


plt.rcParams.update(plt.rcParamsDefault)
def plot_multiple_fit(datasets, labels, scale=1):
    """
    Plots the PCA explained variance and cumulative variance for multiple datasets.

    Parameters:
    datasets (list of np.array): List of datasets to perform PCA on.
    labels (list of str): List of labels corresponding to each dataset.
    scale (float): Scale factor for title.
    """
    plt.rcParams.update({'font.size': 16})
    plt.rc('font', family='Times New Roman')
    plt.rcParams['mathtext.fontset'] = 'cm'
    plt.figure(figsize=(14, 4))
    #plt.rcParams.update({'font.size': 18, 'font.family': 'Times New Roman'})
    
    fit_pca = np.zeros((len(datasets), datasets[0].shape[1]))
    colors = plt.cm.get_cmap('tab10', len(datasets))
    markers = ['o', '+', 's', '^', '*']
    colors = ['black', 'purple', 'gray', 'blue', 'pink']
    for i, (data, marker, color) in enumerate(zip(datasets, markers, colors)):
        X_scaled = robust_scale(data)
        pca = PCA()
        X_pca = pca.fit_transform(X_scaled)
        
        explained_variance_ratio = pca.explained_variance_ratio_
        cumulative_variance_ratio = explained_variance_ratio.cumsum()
        fit_pca[i, :] = cumulative_variance_ratio
        x = np.array([i+1 for i in range(fit_pca.shape[1])]).astype(np.float32)
        y = fit_pca[i, ...].astype(np.float32)
        print(f'{labels[i]}', np.sum(np.where(y <= 0.95, 1, 0)))
        def double_exp_func(x, A, B, k1, k2, x0):
            return A * (1 - np.exp(-k1 * (x - x0))) + B * (1 - np.exp(-k2 * (x - x0)))

        # Fit the Double Exponential function
        popt_double_exp, _ = curve_fit(double_exp_func, x, y, bounds=(0, np.inf))
        y_double_exp = double_exp_func(x, *popt_double_exp)
        r2_double_exp = r2_score(y, y_double_exp)
        print(f"R-squared for double-exp function: {r2_double_exp}")
        
        # Plot the data and fits
        plt.scatter(x, y, color=color, marker=marker, label=f'{labels[i]} R^2: {r2_double_exp}')
        plt.plot(x, y_double_exp, color=color, linestyle='--') # label=f'{labels[i]} - Fit (R²={r2_double_exp:.2f})')

    plt.axhline(y=0.95, color='red', linestyle='dotted')
    plt.ylabel('Explained Variance Ratio', fontsize=22)
    plt.xlabel('Principal Components', fontsize=22)
    plt.yticks(np.arange(.2, 1.05, 0.05))
    plt.legend(loc='best', fontsize=18)
    plt.title(rf'Cumulative Explained Excitation Variance at $l$={scale + 1}', fontsize=28)
    plt.grid(True, alpha=0.5)
    plt.tight_layout()
    plt.savefig(f'explained_var_busi_{scale}.png', dpi=500)
    plt.show()
    
    return fit_pca

# Example usage
scale = 2
labels = ['SE', 'SE-NS', 'SRM', 'FCA', 'Mono']
datasets = [np.array([x_[scale] for x_ in x]).squeeze() for x in [ses, ses_ns, srms, fcas, ssrs]]
fit_pca = plot_multiple_fit(datasets, labels, scale=scale)
```

# The Excitation PCA Threshold (EPT)

By collecting the number of PCs needed to explain 95% of the variance we get the graphs of the EPT for each encoder depth.

We discard FCA because it has no variance for certain scales/datasets.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 24})
scale = 0
# Data for ISIC-18
scales_isic = [1, 2, 3]
se_isic = [1, 3, 4]
srm_isic = [9, 17, 44]
fca_isic = [2, 12, np.inf]
mono_isic = [6, 11, 14]
se_ns_isic = [3, 5, 4]  # New data
models_isic = [se_isic, se_ns_isic, srm_isic, mono_isic]

# Data for Kvasir-SEG
scales_kvasir = [1, 2, 3]
se_kvasir = [4, 4, 8]
srm_kvasir = [9, 24, 28]
fca_kvasir = [10, 5, np.inf]
mono_kvasir = [9, 12, 21]
se_ns_kvasir = [3, 6, 3]  # New data
models_kvasir = [se_kvasir, se_ns_kvasir, srm_kvasir, mono_kvasir, ]

# Data for BUSI
scales_busi = [1, 2, 3]
se_busi = [2, 3, 4]
srm_busi = [8, 12, 13]
fca_busi = [15, 7, np.inf]
mono_busi = [7, 12, 18]
se_ns_busi = [3, 6, 3]  # New data
models_busi = [se_busi, se_ns_busi, srm_busi,  mono_busi, ]

layer_size = scales_isic
labels = ['cSE',  'cSE-NS', 'SRM',  'Mono', ]  # New label
markers = ['o', 'o', 'x', '^']  # New marker
colors = ['black', 'gray', 'purple', 'blue']  # New color

fig, axes = plt.subplots(1, 3, figsize=(27, 6))

# Plot for ISIC-18
ax = axes[0]
for i in range(len(models_isic)):
    ax.plot(layer_size, models_isic[i], color=colors[i], marker=markers[i], linestyle='--', label=f'{labels[i]}')
    ax.axhline(2, color='r', linestyle=':')
ax.set_xticks(layer_size)
ax.legend(loc='best', fontsize=25, ncol=2)
ax.set_title(r'Excitation PCA Threshold ISIC18', fontsize=32)
ax.set_ylabel('N. of Principle Components', fontsize=30)
ax.set_xlabel(r'Depth $l$', fontsize=30)
ax.grid(True, alpha=0.5)
yticks =np.arange(0, 30, 10)
yticks = sorted(set(yticks).union({2}))
ax.set_yticks(yticks)

# Plot for Kvasir-SEG
ax = axes[1]
for i in range(len(models_kvasir)):
    ax.plot(layer_size, models_kvasir[i], color=colors[i], marker=markers[i], linestyle='--', label=f'{labels[i]}')
    ax.axhline(2, color='r', linestyle=':')
ax.set_xticks(layer_size)
ax.legend(loc='best', fontsize=25, ncol=2)
ax.set_title(r'Excitation PCA Threshold Kvasir-SEG', fontsize=32)
ax.set_ylabel('N. of Principle Components', fontsize=30)
ax.set_xlabel(r'Depth $l$', fontsize=30)
ax.grid(True, alpha=0.5)
yticks =np.arange(0, 30, 10)
yticks = sorted(set(yticks).union({2}))
ax.set_yticks(yticks)

# Plot for BUSI
ax = axes[2]
for i in range(len(models_busi)):
    ax.plot(layer_size, models_busi[i], color=colors[i], marker=markers[i], linestyle='--', label=f'{labels[i]}')
    ax.axhline(2, color='r', linestyle=':')
ax.set_xticks(layer_size)
ax.legend(loc='best', fontsize=25, ncol=2)
ax.set_title(r'Excitation PCA Threshold BUSI', fontsize=32)
ax.set_ylabel('N. of Principle Components', fontsize=30)
ax.set_xlabel(r'Depth $l$', fontsize=30)
ax.grid(True, alpha=0.5)
yticks =np.arange(0, 30, 10)
yticks = sorted(set(yticks).union({2}))
ax.set_yticks(yticks)

plt.tight_layout()
# plt.savefig(f'excitation_threshold.png', dpi=500)
plt.show()
```
